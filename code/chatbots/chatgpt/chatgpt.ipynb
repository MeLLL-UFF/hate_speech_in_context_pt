```python
import pandas as pd
from openai import OpenAI
import time
```

```python
client = OpenAI(api_key='API_KEY')
```

```python
type = 'type' #zero_shot; #few_shot...
output_path = f'''output_path/{type}.csv'''
dataset_path = "dataset_path/"
prompt_path = "prompt_path/"
```

```python
base_prompt = ''
with open(prompt_path, 'r') as file:
    base_prompt = file.read()

print(base_prompt)
```

```python
df = pd.read_csv(dataset_path)

df.head()
```

```python
answers = []

i = 0

while i < len(df):
  try:
    prompt = base_prompt +  f'''\nTexto: {df["text"].iloc[i]}\nClasse: '''

    asw = client.completions.create(
            model="gpt-3.5-turbo-instruct",
            prompt=prompt,
            temperature = 0.1,
            max_tokens=20
    )

    i += 1

    aux = asw.choices[0].text
    answers.append(aux.lower())

    time.sleep(2)

  except Exception as e:
    print(f"Error: {e}. Trying again...")
    time.sleep(3)
```

```python
df_answers = pd.DataFrame({
    "text": df['text'].values,
    "answer": answers,
})

df_answers.to_csv(output_path, index=False)
```
