```python
import pandas as pd
import maritalk as mtalk
import time
```

```python
model = mtalk.MariTalk(key="API_KEY")
```

```python
type = 'type' #zero_shot; #few_shot...
output_path = f'''output_path/{type}.csv'''
dataset_path = "dataset_path/"
prompt_path = "prompt_path/"
```

```python
base_prompt = ''
with open(prompt_path, 'r') as file:
    base_prompt = file.read()

print(base_prompt)
```

```python
df = pd.read_csv(dataset_path)


print(len(df))
print()
df.head()
```

```python
answers = []

i = 0

while i < len(df):
  try:
    prompt = base_prompt +  f'''\nTexto: {df["text"].iloc[i]}\nClasse: '''

    asw = model.generate(
              prompt,
              chat_mode=False,
              do_sample=False,
              max_tokens=30,
              temperature=0.3,
              stopping_tokens=["\n"]
        )

    i += 1
    asw = asw.replace("\n", '')
    answers.append(asw.lower())

    time.sleep(4)

  except Exception as e:
    print(f"Error: {e}. Trying again...")
    time.sleep(6)
```

```python
df_answers = pd.DataFrame({
    "text": df['text'].values,
    "answer": answers,
})

df_answers.to_csv(output_path, index=False)
```
